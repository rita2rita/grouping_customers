Нашей компании нужно сгруппировать клиентов для АБ-тестов. Алгоритм группировки очень простой - взять ID клиента (состоит из 5-7 цифр, например 7412567) и найти сумму всех его цифр. Получившееся число и является номером группы, в которую входит данный клиент.

Для того чтобы понять, насколько хорош такой простой алгоритм, тебе нужно написать следующие диагностические функции:

1. Функция, которая подсчитывает число покупателей, попадающих в каждую группу, если нумерация ID сквозная и начинается с 0. На вход функция получает целое число n_customers (количество клиентов).
2. Функция, аналогичная первой, если ID начинается с произвольного числа. На вход функция получает целые числа: n_customers (количество клиентов) и n_first_id (первый ID в последовательности).

```python
def get_customers_groups_id_from_0(n_customers: int):
    groups_dict = {}
    for id_number in range(0, n_customers + 1):
        group_number = 0
        while id_number > 0:
            digit = id_number % 10
            group_number += digit
            id_number = id_number // 10
        if group_number in groups_dict:
            groups_dict[group_number] += 1
        else:
            groups_dict[group_number] = 1
    return groups_dict
```

```python
def get_customers_groups(n_customers: int, n_first_id: int):
    groups_dict = {}
    stop = n_first_id + n_customers + 1
    for id_number in range(n_first_id, stop):
        group_number = 0
        while id_number > 0:
            digit = id_number % 10
            group_number += digit
            id_number = id_number // 10
        if group_number in groups_dict:
            groups_dict[group_number] += 1
        else:
            groups_dict[group_number] = 1
    return groups_dict
```

```python
import matplotlib.pyplot as plt
```

```python
dct = get_customers_groups_id_from_0(1000)
fig, ax = plt.subplots()

x = dct.keys()
y = dct.values()

ax.set_title("Распределение покупателей по группам", fontsize=20)
ax.set_xlabel("Номера групп")
ax.set_ylabel("Кол-во покупателей")

ax.plot(x, y)
ax.bar(dct.keys(), dct.values(), color='grey')
```
